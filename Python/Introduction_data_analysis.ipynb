In the original data we have "?" that hinder our analysis, then we must replace it with 0 or NAN

```python
# For replace "?" to  NaN
import numpy as np
data.replace('?', np.NaN, inplace=True)
```

```python
data
```

```python
data.info()
```

Note that the Non-Null Count changes!

Them note that normalized-losses has  164 non-null   and is the type=object , but we have numbers...
we need to "int" type.


There are several methods to replace a nan or empty space, in cases where there are few we can use the average or the most frequent object.In this case it is preferable to use 0

```python
data['normalized-losses'].fillna(0, inplace=True)
```

in the case of price,peak-rpm,horsepower,stroke, bore and num-of-doors, we use the average or the most frequent object.



For the number of doors, we use the frequency because we dont have "2" but "two". If you prefer is possible change for int. 

```python
data['num-of-doors'].describe()
```

```python
data['num-of-doors'].replace( np.NaN, data['num-of-doors'].describe().freq, inplace=True)
```
